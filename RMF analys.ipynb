```python
df2 = df1.query('Quantity > 0')
df2
```

```python
df2 = df2.query('UnitPrice < 5000')
df2 = df2.query('Quantity < 10000')
```

```python
df2['total_sum'] = df2.UnitPrice*df2.Quantity
```

```python
df2
```

```python
df_abc = df2.groupby('CustomerID', as_index = False).agg({'total_sum': 'sum'}).sort_values('total_sum', ascending = False)
```

```python
df_abc
```

```python
total_sum = df_abc.total_sum.sum()
total_sum
```

```python
df_abc['share_of_total']=df_abc.total_sum/total_sum*100
```

```python
df_abc['cum_sum']=df_abc.share_of_total.cumsum()
df_abc
```

```python
df_abc['abc'] = np.where(df_abc['cum_sum']<80, 'A', np.where((df_abc['cum_sum']<95), 'B', 'C'))
df_abc
```

```python
df_abc.query('abc == "A"')
```
